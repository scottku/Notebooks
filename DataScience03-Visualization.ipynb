```python
# mtcars 데이터셋을 이용한 산점도 그래프(Scatter plot)
import pandas as pd
mtcars = pd.read_csv("./data/mtcars.csv") # 헤더가 있고, 구분자는 콤마
mtcars
```

```python
# EDA
print("Size: ", mtcars.shape) # 레코드는 32개, column 수는 12개
# 위쪽 데이터 확인
mtcars.head()
mtcars.tail()
```

```python
mtcars.describe()
```

```python
plt.scatter(mtcars['wt'], mtcars['mpg'])
#연비와 중량 산점도
plt.xlabel('Weight')
plt.ylabel('Mile per Gallon')
plt.title("Weight vs MpG")
```

```python
# 상관 계수
print("상관계수(wt vs mpg)")
mtcars.corr() # 상관계수 - 모든 수치 변수들 사이의 상관 계수
mtcars['wt'].corr(mtcars['mpg']) # wt 변수와 mpg 변수 사이의 상관 계수
# -1 : 역비례, 0 : 관계 없음, 1 : 정비례
```

```python
# 상자 그림(Box Plot)
# plt.boxplot(mtcars['mpg']) : 단일 변수의 상자 그림
plt.boxplot([mtcars['mpg'], mtcars['wt']])
plt.show()

print("mpg의 기초 통계량")
display(mtcars['mpg'].describe())
```

```python
# 분위수 확인
quantiles = mtcars.quantile([0.25, 0.5, 0.75]) # 모든 변수의 분위수
display(quantiles)
```

```python
#mpg의 1, 3사분위
first_quantile = mtcars['mpg'].quantile(0.25)
third_quantile = mtcars['mpg'].quantile(0.75)
print("1사분위 {}, 3사분위 {}".format(first_quantile, third_quantile))

iqr = third_quantile - first_quantile
print("IQR of mpg:", iqr)
```

```python
# mpg의 상단 극단치 경계
# Q3 + IQR * 1.5
border_top = third_quantile + iqr * 1.5
border_top
border_bot = first_quantile - iqr * 1.5
border_bot
```

```python
# mpg가 극단치를 가지는 자동차는?
#  = mpg가 상단 극단치를 초과한 레코드
# mtcars['mpg'] > border_top -> boolean 연산
outliers = mtcars[mtcars['mpg']>border_top]
display(outliers)
```
