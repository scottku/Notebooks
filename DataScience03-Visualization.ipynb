```python
# maplotlib로 차트 그리기
# Anscombe의 시각화 강조를 위한 4분할 그래프
# 통계 수치, 상관관계, 회귀선이 모두 같다면 데이터 모양이 비슷할까?

import seaborn as sns
# 샘플 데이터 set 목록
print(sns.get_dataset_names())
# anscombe 데이터셋 불러오기
anscombe = sns.load_dataset("anscombe")
anscombe
```

```python
# 데이터 서브셋의 갯수 확인
print(anscombe['dataset'].nunique()) # 유니크한 값의 개수
```

```python
ds1 = anscombe[anscombe['dataset']=='I'] # 데이터셋 1
ds2 = anscombe[anscombe['dataset']=='II'] # 데이터셋 2
ds3 = anscombe[anscombe['dataset']=='III'] # 데이터셋 3
ds4 = anscombe[anscombe['dataset']=='IV'] # 데이터셋 4
```

```python
ds1.describe() # 데이터셋 1의 기초 통계량
```

```python
ds2.describe()
```

```python
# dataset1의 그래프
import matplotlib.pyplot as plt
plt.plot(ds1['x'], ds1['y'], 'o')
```

```python
# dataset2의 그래프
plt.plot(ds2['x'], ds2['y'], "^", color="orange")
```

```python
# 앤스콤 4분할 그래프 그리기
# 그래프의 기본 툴 = figure
fig = plt.figure(figsize=(12, 12))
# 기본틀에 격자 추가
axes1 = fig.add_subplot(2,2,1) # 2 x 2 격자의 1번 영역
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)

# 격자에 그래프 그리기
axes1.plot(ds1['x'], ds1['y'], 'o')
axes2.plot(ds2['x'], ds2['y'], '+', color="green")
# 서브플롯에 타이틀 출력
axes1.plot(ds1['x'], ds2['y'], '+')
# plt.show() # 그래프 출력

fig.suptitle("Anscombe Datasets")
# 레이아웃 조정
fig.tight_layout()

plt.show()
```
